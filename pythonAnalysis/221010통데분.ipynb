```python
%%time
import time
#Mean 내부 함수 사용
print(np.mean(df["body_mass_g"]))
print(np.mean(df["bill_length_mm"]))
```

```python
%%time
#Mean 내부 함수 사용 X

length =0 
for val in df["body_mass_g"]:
    length=length+val
    
length
```

```python
%%time
#Median 내부 함수
print(df["body_mass_g"].median())
print(df["bill_length_mm"].median())
```

```python
%%time
#Median 내부함수X
def get_median(data):
    median=0
    idx=0
    data=sorted(data)
    if (len(data)%2==0):
        idx=len(data)//2
        median=(data[idx-1]+data[idx])/2
    else:
        idx=len(data)//2
        median=data[idx]
    return median
    
print(get_median(df["body_mass_g"]))
print(get_median(df["bill_length_mm"]))
```

```python
%%time
#Variance 내부 함수
print(np.var(df["body_mass_g"]))
print(np.var(df["bill_length_mm"]))
```

```python
%%time
#Variance 내부 함수X

def variance(data):
    m=sum(data)/len(data)
    ss=0
    for i in data:
        ss+=pow(i-m,2)
# 분산 : 편차 제곱 합 / 데이터 개수
    var=ss/len(data)
    return var

print(variance(df["body_mass_g"]))
print(variance(df["bill_length_mm"]))
```

```python
%%time
#Mad 내부 함수
print(df["bill_length_mm"].mad())
print(df["body_mass_g"].mad())
```

```python
%%time
#Mad 내부 함수X
def get_MAD(data):
    return np.median(np.absolute(data - np.median(data, axis=0)), axis=0)

print(get_MAD(df["body_mass_g"]))
print(get_MAD(df["bill_length_mm"]))
```

```python
%%time
#왜도 내장함수
from scipy.stats import skew
print(df["body_mass_g"].skew())
print(df["bill_length_mm"].skew())
```

```python
%%time
#왜도 내장함수X
def get_skewness(inp):
    len_inp = len(inp)
    result = 0
    for i in inp:
        result += ((i - np.mean(inp)) / np.std(inp)) ** 3
    result = result / len_inp
    return result

print(get_skewness(df["body_mass_g"]))
print(get_skewness(df["bill_length_mm"]))
```

```python
%%time
from scipy.stats import kurtosis
# 첨도 내장함수
print(df["bill_length_mm"].kurtosis())
print(df["body_mass_g"].kurtosis())
```

```python
%%time
# 첨도 내장함수X
def get_kurtosis(inp):
    len_inp = len(inp)
    result = 0
    for i in inp:
        result += ((i - np.mean(inp)) / np.std(inp)) ** 4
    result = (result / len_inp) - 3
    return result
print(get_kurtosis(df["body_mass_g"]))
print(get_kurtosis(df["bill_length_mm"]))
```

```python
%%time
#IQR 내장 함수
print(df["bill_length_mm"].quantile((0.25, 0.50, 0.75))) 

print()
print(df["body_mass_g"].quantile((0.25, 0.50, 0.75))) 
```

```python
%%time
#IQR 내장 함수X
def quartile(data, p):
    n=len(data)+1
    data=sorted(data)
    diff=n*p-int(n*p)
    if diff==0:
        return data[int(n*p-1)]
    else:
        return data[int(n*p-1)]*(1-diff)+data[int(n*p)]*diff
    
print(quartile(df["bill_length_mm"],0.25))
print(quartile(df["bill_length_mm"],0.50))
print(quartile(df["bill_length_mm"],0.75))

print()
print(quartile(df["body_mass_g"],0.25))
print(quartile(df["body_mass_g"],0.50))
print(quartile(df["body_mass_g"],0.75))
```
