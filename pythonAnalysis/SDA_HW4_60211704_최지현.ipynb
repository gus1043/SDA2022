```python
import pandas as pd
import seaborn as sns
import numpy as np
from scipy import stats
```

```python
data = pd.read_csv('student-mat.csv')
```

```python
data
```

```python
#Q1. 성별에 따른 최종 성적이 통계적으로 유의미하게 차이가 나는가
data_m=data[data['sex'] == 'M']
data_w=data[data['sex'] == 'F']
print(data_m["G3"].mean())
print(data_w["G3"].mean())
print(data_m[data_m["G3"]>0]["G3"].mean())
print(data_w[data_w["G3"]>0]["G3"].mean())
```

```python
#t-test를 위한 조건 체크1 : Independence
#p-value>0.05면 등분산성 만족
#p-value=0.7274377618424044이므로 등분산성이 있다
from scipy.stats import bartlett
stat, p =bartlett(data_m[data_m["G3"]>0]["G3"], data_w[data_w["G3"]>0]["G3"])
print(stat,p)
#t-test를 위한 조건 체크2 : Normality
#p-value>0.05면 전규성 만족
#data_m p_value = 0.009935521520674229, data_w p-value=0.006041768938302994이므로 정규성을 따르지 않는다.
from scipy.stats import shapiro
stat_m, p=shapiro(data_m[data_m["G3"]>0]["G3"])
print(stat_m, p)
stat_w, p=shapiro(data_w[data_w["G3"]>0]["G3"])
print(stat_w, p)
```

```python
#정규성을 따르지 않으므로 비모수 검증 필요
stats.ranksums(data_m[data_m["G3"]>0]["G3"], data_w[data_w["G3"]>0]["G3"])
```
