```python
import requests
from bs4 import BeautifulSoup
import time
import csv
import pandas as pd
data=pd.read_csv('ALL_KC_KOBIS_BOX_OFFIC_MOVIE_INFO.csv', index_col=0)

data_all=data[(data['GRAD_NM']!='청소년관람불가') & (data['NLTY_NM']=='한국')]
print(data_all['MOVIE_NM'].duplicated().value_counts())
data_all
```

```python
data_list = list()
for row in data_all['MOVIE_NM']:
    data_list.append(row)
```

```python
from selenium import webdriver
import time
```

```python
path = "C:/chromedriver.exe"
driver = webdriver.Chrome(path)

driver.get("https://movie.naver.com/")
time.sleep(3)
```

```python
import openpyxl
# Workbook 생성
wb = openpyxl.Workbook()

# Sheet 활성
sheet = wb.active

#데이터프레임 내 header(변수명)생성
sheet.append(["NO", "제목", "url", "평론가 평점", "네티즌 평점"])
```

```python
#url 따오기
no=1
for i in data_list: 
    text = i

    driver.find_element_by_id("ipt_tx_srch").click()

    element = driver.find_element_by_id("ipt_tx_srch")

    element.send_keys(text)

    driver.find_element_by_css_selector("button.btn_srch").click()

    driver.find_element_by_css_selector("ul.search_list_1 > li > p > a ").click()
    data_all['url']=driver.current_url
    sheet.append([no, i, driver.current_url])
    no+=1
wb.save('movie_url.xlsx')
```

```python
#크롤링
fpath = 'movie_url.xlsx'
wb = openpyxl.load_workbook(fpath)
ws = wb['Sheet'] # wb.active는 현재 활성화된 창

url=pd.read_excel('movie_url.xlsx', index_col=0)
url=url['url']

import re
for i in range(1,len(url)+1):
    html=requests.get(url[i])
    bs_html = BeautifulSoup(html.content,"html.parser")
    critic_review = bs_html.find("a",{"class":"spc"})
    netizen_review = bs_html.find("a",{"id":"pointNetizenPersentBasic"})
    
    print("순서 :", i)

    if critic_review is not None:  
        print("평론가 평점: ", float(re.findall('\d+.\d+',critic_review.text)[0]))
        ws['D'+str(i+1)]=float(re.findall('\d+.\d+',critic_review.text)[0])
    else: 
        print("평론가 평점 없음")

    if netizen_review is not None: 
        print("네티즌 평점: ",netizen_review.text)
        ws['E'+str(i+1)]=netizen_review.text
    else : 
        print("네티즌 평점 없음")
wb.save(fpath)
```
